## Importación de Librerías

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

## Lectura del dataset

```python
# Ruta del archivo 
file_path = 'attachments/Conjunto_servidores_p_blicos_20240819.csv'
# Cargar el archivo 
df = pd.read_csv(file_path)
```

## Filtrar las columnas relevantes 

```python
df = df[['Sexo', 'Meses de Experiencia Público', 'Meses de Experiencia Privado',
        'Nivel Educativo', 'Tipo de Nombramiento', 'Nivel Jerarquico Empleo',
        'Dependencia Empleo Actual', 'Asignación Básica Salarial']]

# Mostrar las primeras 5 filas del DataFrame
df.head()
```

## Cambiar los valores Nulos dentro de las columnas por 0 o 'desconocido' si son categoricas

```python
# Convertir las columnas numéricas y rellenar nulos
df.loc[:, 'Meses de Experiencia Público'] = pd.to_numeric(df['Meses de Experiencia Público'], errors='coerce').fillna(0)
df.loc[:, 'Meses de Experiencia Privado'] = pd.to_numeric(df['Meses de Experiencia Privado'], errors='coerce').fillna(0)

# Rellenar las columnas categóricas con 'desconocido'
df.loc[:, 'Nivel Educativo'] = df['Nivel Educativo'].fillna('desconocido')
df.loc[:, 'Nivel Jerarquico Empleo'] = df['Nivel Jerarquico Empleo'].fillna('desconocido')
df.loc[:, 'Dependencia Empleo Actual'] = df['Dependencia Empleo Actual'].fillna('desconocido')

# Mostrar las primeras 5 filas del DataFrame
df.head()
```

## Eliminar filas donde es nulo en la columna asignacion salarial
## Convertir la columna de asignacion salarial a numerico

```python
# Eliminar filas donde es nulo
df = df.dropna(subset=['Asignación Básica Salarial'])

# Convertir a numérica, eliminando caracteres especiales
df['Asignación Básica Salarial'] = df['Asignación Básica Salarial'].replace({'$': '', ',': ''}, regex=True).astype(float)
```

## Convertir las columnas categoricas a numericas con la libreria LabelEncoder

```python
categorical_columns = ['Sexo', 'Nivel Educativo', 'Tipo de Nombramiento', 'Nivel Jerarquico Empleo', 'Dependencia Empleo Actual']
label_encoders = {}

for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))
    label_encoders[column] = le

#Visualizar cuales son las enumeraciones dependiendo la columna
#for column, le in label_encoders.items():
    #print(f"Columna: {column}")
    #for class_index, class_label in enumerate(le.classes_):
        #print(f"{class_label} -> {class_index}")
    #print("\n")

#Visualizar cuales son las enumeraciones dependiendo la columna
for column, le in label_encoders.items():
    print(f"Columna: {column}")
    for class_index, class_label in enumerate(le.classes_):
        print(f"{class_label} -> {class_index}")
    print("\n")
    
```

# Guardar el DataFrame procesado en un archivo CSV

```python
df.to_csv('attachments/procesado.csv', index=False, encoding='utf-8')
```
